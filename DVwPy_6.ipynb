## ***SEABORN-PART3***

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

### ***COMPARÄ°SON PLOTS***

# **Catplot**

```python
titanic = sns.load_dataset("titanic")
titanic.head(1)
```

```python
sns.catplot(x="pclass", y="survived", data = titanic, kind = "bar", estimator=np.count_nonzero);
```

```python
sns.catplot(x="pclass", y="survived", data = titanic, kind = "bar", estimator=np.sum, ci ="sd");
```

```python
sns.catplot(x="pclass", y="survived", data = titanic, kind = "bar", estimator=np.sum, hue ="sex", col = "alone", row="embarked");
```

```python
sns.catplot(x="pclass", y="survived", data = titanic, kind = "bar", estimator=np.sum, col = "sex");
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.catplot(x="day", y="total_bill", data = tips, kind= "swarm", hue="smoker");
```

# **Jointplot**

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind = "reg"); # kind = "reg", "hex", "kde", "scatter"
 
```

# **Pairplot**

```python
sns.pairplot(tips, hue= "sex", palette = "viridis", corner = True, diag_kind="hist"); #pairplotta x ve y yok. diag_kind= "auto", "hist", "kde", None
```

# **PairGrid**

```python
g = sns.PairGrid(tips)

g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot, color = "r")
```

```python
g = sns.PairGrid(tips, hue ="sex", palette="viridis")

g = g.map_upper(sns.scatterplot, linewidth=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();
```

```python
g = sns.pairplot(tips)
g.map_lower(sns.kdeplot, color='purple')
g.map_upper(sns.kdeplot, color='g');
```

# **FacetGrid**

```python
g = sns.FacetGrid(data = tips, col="time", row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(data = tips, col="time", row="smoker")
g = g.map(plt.scatter, "total_bill", "tip")
```

```python
g = sns.FacetGrid(data = tips, col="time", row="smoker")
g = g.map(plt.scatter, "total_bill", "tip", color="r")
```
